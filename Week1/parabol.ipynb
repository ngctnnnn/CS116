```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
from time import sleep
```

```python
x = np.array(range(-10, 11))
y = x**2
```

```python
plt.plot(x, y)
for i in x:
    plt.plot(i, i**2, 'ro')
    display.display(plt.gcf())
    display.clear_output(wait=True)
    sleep(0.5)
plt.show()
```
